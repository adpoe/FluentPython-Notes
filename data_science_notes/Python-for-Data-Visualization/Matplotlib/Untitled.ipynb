```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[0].set_title('Second Plot')

plt.tight_layout()
```

```python
fig, axes = plt.subplots()
axes.plot(x,x**2, label='x sq')
axes.plot(x,x**3, label='x cubed')

axes.legend()

axes.legend(loc=0)
```
